```python
#Q1 Data Visualization and Statistical Measures: 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df1=pd.read_csv("iris.csv")
```

```python
df1
```

```python
#(i) Scatter graph
plt.scatter(df1['Sepal.Length'],df1['Sepal.Width'],df1['Petal.Length'],df1['Petal.Width'])
```

```python
plt.scatter(df1['Sepal.Length'],df1['Petal.Width'],alpha=0.7)
```

```python
#(ii) hisplot
sns.histplot(df1['Sepal.Length'],kde=True,bins=20)
```

```python
sns.histplot(df1['Petal.Length'],kde=True,bins=20)
```

```python
#(iii) boxplot
sns.boxplot(x=df1['Sepal.Length'],y=df1['Sepal.Width'])
```

```python
sns.boxplot(x=df1['Petal.Length'],y=df1['Petal.Width'])
```

```python
#(iv) countplot
sns.countplot(x='Species', data=df1)
plt.show()
```
